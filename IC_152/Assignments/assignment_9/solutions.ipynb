## Question 1: Reproducible Random Integers
Create 10 random integers using numpy.random.randint. Now create the same set of random integers again. (Hint: use seed)

### Algorithm for Question 1
1. Import numpy as np.
2. Set the random seed using np.random.seed() to ensure reproducibility.
3. Generate 10 random integers using np.random.randint().
4. Set the same seed again and generate the 10 random integers to confirm they are identical.

```python
# Solution for Question 1
import numpy as np
np.random.seed(42)
rand_ints1 = np.random.randint(0, 100, 10)
print("First set:", rand_ints1)
np.random.seed(42)
rand_ints2 = np.random.randint(0, 100, 10)
print("Second set:", rand_ints2)
```

### Logic Explanation
- Setting the same seed before generating random numbers ensures the output is reproducible and identical each time.

## Question 2: Reproducible Random Floats with Different Seeds
Create 20 random floating point numbers using numpy.random.normal. Now create the same set of random floating point numbers again. Repeat this exercise to generate a different set of 20 random floating point numbers, twice. (Hint: use different seeds for each repetition of numbers)

### Algorithm for Question 2
1. Import numpy as np.
2. Set a seed and generate 20 random floats using np.random.normal().
3. Set the same seed and generate the numbers again to confirm they match.
4. Change the seed and generate a new set of 20 random floats. Repeat for another seed.

```python
# Solution for Question 2
np.random.seed(1)
floats1 = np.random.normal(size=20)
print("First set (seed=1):", floats1)
np.random.seed(1)
floats2 = np.random.normal(size=20)
print("Second set (seed=1):", floats2)
np.random.seed(2)
floats3 = np.random.normal(size=20)
print("Different set (seed=2):", floats3)
np.random.seed(3)
floats4 = np.random.normal(size=20)
print("Different set (seed=3):", floats4)
```

### Logic Explanation
- Using the same seed produces the same sequence of random numbers.
- Changing the seed produces a different sequence.

## Question 3: Plot Sin2x, Cos2x, and Their Sum
Plot Sin2x, Cos2x, and their sum in the same plot for 1000 common values of x ranging between -10 to 10 radians. Give adequate title, x-labels, y-labels and legends. Use a font size of 24 for all the text in your plot. (Using subplot is not allowed in this question).

### Algorithm for Question 3
1. Import numpy and matplotlib.pyplot.
2. Generate 1000 values of x between -10 and 10.
3. Compute sin(2x), cos(2x), and their sum for all x.
4. Plot all three on the same figure with appropriate labels, title, and legend, using font size 24.

```python
# Solution for Question 3
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 1000)
sinx_2 = np.sin(x**2)
cosx_2 = np.cos(x**2)
sumx_2 = sinx_2 + cosx_2
plt.figure(figsize=(12, 6))
plt.plot(x, sinx_2, label='sin(2x)')
plt.plot(x, cosx_2, label='cos(2x)')
plt.plot(x, sumx_2, label='sin(2x) + cos(2x)')
plt.title('sin(2x), cos(2x), and their sum', fontsize=24)
plt.xlabel('x (radians)', fontsize=24)
plt.ylabel('Value', fontsize=24)
plt.legend(fontsize=24)
plt.tick_params(labelsize=18)
plt.show()
```

### Logic Explanation
- All three functions are plotted on the same axes for direct comparison, with large font sizes for clarity.

## Question 4: Mean and Median as Minimizers
Generate 100 random integers. Find their mean and median using numpy.mean() and numpy.median(). In the same figure write python code to create different subplots for:
- a. Graphically showing that mean is minimizer of (x - xi)^2
- b. Graphically showing that median is minimizer of |x - xi|

### Algorithm for Question 4
1. Generate 100 random integers.
2. Compute the mean and median.
3. For a range of x values, compute the sum of squared differences and sum of absolute differences.
4. Plot both minimization graphs in subplots, marking the mean and median.

```python
# Solution for Question 4
rand_ints = np.random.randint(0, 100, 100)
mean_val = np.mean(rand_ints)
median_val = np.median(rand_ints)
x_range = np.linspace(min(rand_ints), max(rand_ints), 500)
sq_sum = [np.sum((x - rand_ints)**2) for x in x_range]
abs_sum = [np.sum(np.abs(x - rand_ints)) for x in x_range]
fig, axs = plt.subplots(1, 2, figsize=(16, 6))
axs[0].plot(x_range, sq_sum)
axs[0].axvline(mean_val, color='r', linestyle='--', label='Mean')
axs[0].set_title('Mean minimizes sum of squared differences')
axs[0].legend()
axs[1].plot(x_range, abs_sum)
axs[1].axvline(median_val, color='g', linestyle='--', label='Median')
axs[1].set_title('Median minimizes sum of absolute differences')
axs[1].legend()
plt.show()
```

### Logic Explanation
- The left subplot shows the sum of squared differences minimized at the mean.
- The right subplot shows the sum of absolute differences minimized at the median.

## Question 5: Statistical Regularity and Central Limit Theorem
### a) Simulate picking a ball at random (0, 1, 2, 3) for 50, 500, 5000, and 50000 trials. Plot the relative frequency of each observed value vs number of trials. Save the plots as required.

### Algorithm for Question 5a
1. For each trial count (50, 500, 5000, 50000):
    a. Generate random integers between 0 and 3 for the given number of trials.
    b. Count the occurrences of each value.
    c. Compute relative frequencies.
    d. Plot and save the figure with the specified filename.

```python
# Solution for Question 5a
import random
trial_counts = [50, 500, 5000, 50000]
for n in trial_counts:
    results = [random.randint(0, 3) for _ in range(n)]
    rel_freq = [results.count(i)/n for i in range(4)]
    plt.figure()
    plt.bar(range(4), rel_freq)
    plt.xlabel('Ball Number')
    plt.ylabel('Relative Frequency')
    plt.title(f'Relative Frequency for {n} Trials')
    plt.ylim(0, 1)
    plt.savefig(f'problem5a_{n}.png')
    plt.close()
```

### Logic Explanation
- As the number of trials increases, the relative frequencies approach the theoretical probability (0.25 for each ball).

### b) Central Limit Theorem: Distribution of Sample Means
Plot the histograms of sample means (observed values) of 5 trials, by repeating it 50000 times and verify. Save the plot as problem5b_5_50000.png. Run the code with 50 trials instead of 5 in each repetition of 50000. Save the plot as problem5b_50_50000.png.

### Algorithm for Question 5b
1. For each sample size (5 and 50):
    a. Repeat 50000 times:
        i. Generate random integers between 0 and 3 for the sample size.
        ii. Compute the mean of the sample.
    b. Plot the histogram of sample means and save the figure.

```python
# Solution for Question 5b
sample_sizes = [5, 50]
for size in sample_sizes:
    means = []
    for _ in range(50000):
        sample = [random.randint(0, 3) for _ in range(size)]
        means.append(np.mean(sample))
    plt.figure()
    plt.hist(means, bins=30, edgecolor='black')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of Sample Means (n={size}, 50000 repetitions)')
    plt.savefig(f'problem5b_{size}_50000.png')
    plt.close()
```

### Logic Explanation
- As the sample size increases, the distribution of sample means becomes more concentrated and approaches a normal distribution, illustrating the Central Limit Theorem.
